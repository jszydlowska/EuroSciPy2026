```python
import pandas as pd
import polars as pl

pd_df = pd.read_csv('palmerpenguins_original.csv')
pl_df = pl.read_csv('palmerpenguins_original.csv')
```

## Inspect

```python
pd_df.head()
```

```python
pd_df.info()
```

```python
pl_df.head()
```

```python
pl_df.schema
```

## Cleaning

```python
pd_df.isna().sum()
```

```python
pd_df = pd_df.dropna()
```

```python
pd_df.isna().sum()
```

```python
pl_df.null_count()
```

```python
pl_df = pl_df.drop_nulls()
```

```python
pl_df.null_count()
```

## Selecting

```python
pd_df[['species','body_mass_g']].head()
```

```python
pl_df.select(['species','body_mass_g']).head()
```

## Filtering

```python
pd_df[pd_df['species']=='Gentoo']
```

```python
pl_df.filter(pl.col('species')=='Gentoo')
```

## Transform

```python
pd_df['body_mass_kg']=pd_df['body_mass_g']/1000
pd_df.head()
```

```python
pl_df=pl_df.with_columns((pl.col('body_mass_g')/1000).alias('body_mass_kg'))
pl_df.head()
```

## Aggregation

```python
pd_df.groupby('species')['body_mass_g'].mean()
```

```python
pl_df.group_by('species').agg(pl.mean('body_mass_g'))
```
